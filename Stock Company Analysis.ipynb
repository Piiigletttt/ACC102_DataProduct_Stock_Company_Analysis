```python
import wrds
import pandas as pd
import matplotlib.pyplot as plt
import getpass

print("Welcome to WRDS Stock Analysis System")
print("Please enter your WRDS account information to continue.\n")

username = input("Please enter your WRDS username: ")
password = getpass.getpass("Please enter your WRDS password: ")

db = wrds.Connection(
    wrds_username=username,
    wrds_password=password
)

print("Connected successfully!")

ticker = input(
    "Please input three companies，separated by commas（e.g.AAPL,MSFT,GOOGL）: "
).split(",")

company_dict = {
    ticker[0].strip(): "Company A",
    ticker[1].strip(): "Company B",
    ticker[2].strip(): "Company C"
}

start_date = "2024-01-01"
stock_tuple = tuple(company_dict.keys())

# PART 1 Stock Monthly Return 
selected_columns = """
b.htsymbol,
b.permno,
a.date,
a.prc,
a.ret
"""

sql_query = f"""
SELECT {selected_columns}
FROM crsp.msf a
LEFT JOIN crsp.msfhdr b
ON a.permno=b.permno
WHERE b.htsymbol IN {stock_tuple}
AND a.date >= '{start_date}'
"""

stock = db.raw_sql(sql_query, date_cols=["date"])

stock = stock.rename(columns={
    "htsymbol": "tic",
    "prc": "price",
    "ret": "monthly_return"
})

stock = stock.sort_values(
    by=["tic", "date"]
).reset_index(drop=True)

# PART 2 Monthly Return Industry Avg
industry_return = stock.groupby("date")["monthly_return"].mean().reset_index()
industry_return["tic"] = "US Industry Avg"

# PART 3 Return Chart
print("\nGenerating Monthly Return Trend Chart...")
print("Tip: Compare each company's return volatility and performance versus industry average.")

plt.figure(figsize=(12,6))

for i in company_dict.keys():
    temp = stock[stock["tic"] == i]

    plt.plot(temp["date"], temp["monthly_return"], marker="o", linewidth=2, label=i)

plt.plot(
    industry_return["date"],
    industry_return["monthly_return"],
    marker="s",
    linestyle="--",
    linewidth=3,
    label="US Industry Avg"
)

plt.title("Monthly Return vs Industry Average")
plt.xlabel("Date")
plt.ylabel("Monthly Return")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# PART 4 Monthly PE / PB / DY / ROE
ratio_query = f"""
SELECT
tic,
datadate,
prcc_f,
csho,
ceq,
ni,
dvc
FROM comp.funda
WHERE tic IN {stock_tuple}
AND datadate >= '{start_date}'
AND indfmt='INDL'
AND datafmt='STD'
AND consol='C'
AND popsrc='D'
ORDER BY tic, datadate
"""

ratio = db.raw_sql(ratio_query, date_cols=["datadate"])

ratio["tic"] = ratio["tic"].astype(str)
ratio["datadate"] = pd.to_datetime(ratio["datadate"])

monthly_list = []

for company in company_dict.keys():

    temp = ratio[ratio["tic"] == company].copy()
    temp = temp.sort_values("datadate").reset_index(drop=True)

    month_range = pd.date_range(
        start=start_date,
        end=pd.Timestamp.today(),
        freq="ME"
    )

    monthly_df = pd.DataFrame({
        "datadate": pd.to_datetime(month_range)
    })

    monthly_df["tic"] = str(company)

    monthly_df["tic"] = monthly_df["tic"].astype(str)
    temp["tic"] = temp["tic"].astype(str)

    monthly_df = pd.merge_asof(
        monthly_df.sort_values("datadate"),
        temp.sort_values("datadate"),
        on="datadate",
        by="tic",
        direction="backward"
    )

    monthly_list.append(monthly_df)

ratio = pd.concat(monthly_list, ignore_index=True)

# Calculate Ratios
ratio["EPS"] = ratio["ni"] / ratio["csho"]
ratio["BVPS"] = ratio["ceq"] / ratio["csho"]

ratio["P/E"] = ratio["prcc_f"] / ratio["EPS"]
ratio["P/B"] = ratio["prcc_f"] / ratio["BVPS"]
ratio["Dividend_Yield"] = ratio["dvc"] / ratio["prcc_f"]
ratio["ROE"] = ratio["ni"] / ratio["ceq"]

ratio = ratio[
    ["tic", "datadate", "P/E", "P/B", "Dividend_Yield", "ROE"]
].round(4)

industry_ratio = ratio.groupby("datadate").agg({
    "P/E": "mean",
    "P/B": "mean",
    "Dividend_Yield": "mean",
    "ROE": "mean"
}).reset_index()

industry_ratio["tic"] = "US Industry Avg"

industry_ratio = industry_ratio[
    ["tic", "datadate", "P/E", "P/B", "Dividend_Yield", "ROE"]
]
ratio_all = pd.concat([ratio, industry_ratio], ignore_index=True)

# Tables
print("\nPE Monthly Table")
display(ratio_all[["tic", "datadate", "P/E"]])

print("\nPB Monthly Table")
display(ratio_all[["tic", "datadate", "P/B"]])

print("\nDividend Yield Monthly Table")
display(ratio_all[["tic", "datadate", "Dividend_Yield"]])

print("\nROE Monthly Table")
display(ratio_all[["tic", "datadate", "ROE"]])

# Charts
# ---------- P/E ----------
print("\nGenerating P/E Trend Chart...")
print("Tip: Higher P/E may indicate growth expectations; compare valuation levels.")

plt.figure(figsize=(12,6))
for i in list(company_dict.keys()) + ["US Industry Avg"]:
    temp = ratio_all[ratio_all["tic"] == i]

    if i == "US Industry Avg":
        plt.plot(
            temp["datadate"], temp["P/E"],
            linestyle="--", linewidth=3, marker="s", label=i
        )
    else:
        plt.plot(
            temp["datadate"], temp["P/E"],
            marker="o", label=i
        )

plt.title("Monthly P/E Trend vs Industry Average")
plt.xlabel("Date")
plt.ylabel("P/E")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------- P/B ----------
print("\nGenerating P/B Trend Chart...")
print("Tip: Lower P/B may suggest undervaluation; compare asset pricing efficiency.")

plt.figure(figsize=(12,6))
for i in list(company_dict.keys()) + ["US Industry Avg"]:
    temp = ratio_all[ratio_all["tic"] == i]

    if i == "US Industry Avg":
        plt.plot(
            temp["datadate"], temp["P/B"],
            linestyle="--", linewidth=3, marker="s", label=i
        )
    else:
        plt.plot(
            temp["datadate"], temp["P/B"],
            marker="o", label=i
        )

plt.title("Monthly P/B Trend vs Industry Average")
plt.xlabel("Date")
plt.ylabel("P/B")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------- Dividend Yield ----------
print("\nGenerating Dividend Yield Trend Chart...")
print("Tip: Higher dividend yield may imply stronger cash return or weaker price momentum.")

plt.figure(figsize=(12,6))
for i in list(company_dict.keys()) + ["US Industry Avg"]:
    temp = ratio_all[ratio_all["tic"] == i]

    if i == "US Industry Avg":
        plt.plot(
            temp["datadate"], temp["Dividend_Yield"],
            linestyle="--", linewidth=3, marker="s", label=i
        )
    else:
        plt.plot(
            temp["datadate"], temp["Dividend_Yield"],
            marker="o", label=i
        )

plt.title("Monthly Dividend Yield Trend vs Industry Average")
plt.xlabel("Date")
plt.ylabel("Dividend Yield")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------- ROE ----------
print("\nGenerating ROE Trend Chart...")
print("Tip: Higher ROE often means stronger profitability and capital efficiency.")

plt.figure(figsize=(12,6))
for i in list(company_dict.keys()) + ["US Industry Avg"]:
    temp = ratio_all[ratio_all["tic"] == i]

    if i == "US Industry Avg":
        plt.plot(
            temp["datadate"], temp["ROE"],
            linestyle="--", linewidth=3, marker="s", label=i
        )
    else:
        plt.plot(
            temp["datadate"], temp["ROE"],
            marker="o", label=i
        )

plt.title("Monthly ROE Trend vs Industry Average")
plt.xlabel("Date")
plt.ylabel("ROE")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# PART 5 Technical Indicators 
print("\nTechnical Indicator Tables")

for stock_name in company_dict.keys():

    tech = stock[stock["tic"] == stock_name].copy()
    tech = tech.sort_values("date")

    tech["MA5"] = tech["price"].rolling(5).mean()
    tech["MA10"] = tech["price"].rolling(10).mean()
    tech["MA20"] = tech["price"].rolling(20).mean()
    tech["MA30"] = tech["price"].rolling(30).mean()

    ema12 = tech["price"].ewm(span=12, adjust=False).mean()
    ema26 = tech["price"].ewm(span=26, adjust=False).mean()

    tech["MACD"] = ema12 - ema26
    tech["Signal"] = tech["MACD"].ewm(span=9, adjust=False).mean()

    tech_table = tech[
        ["tic", "date", "price", "MA5", "MA10", "MA20", "MA30", "MACD"]
    ].round(4)

    print(f"\n{stock_name} Technical Table")
    display(tech_table)

    # MA Chart
    print(f"\nGenerating {stock_name} Moving Average Chart...")
    print("Tip: When short MA crosses above long MA, bullish momentum may appear.")
    plt.figure(figsize=(12,6))
    plt.plot(tech["date"], tech["price"], label="Price", linewidth=2)
    plt.plot(tech["date"], tech["MA5"], label="MA5")
    plt.plot(tech["date"], tech["MA10"], label="MA10")
    plt.plot(tech["date"], tech["MA20"], label="MA20")
    plt.plot(tech["date"], tech["MA30"], label="MA30")
    
    plt.title(f"{stock_name} Moving Average")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # MACD Chart
    print(f"\nGenerating {stock_name} MACD Chart...")
    print("Tip: MACD crossing above Signal may indicate upward momentum.")
    
    plt.figure(figsize=(12,4))
    plt.plot(tech["date"], tech["MACD"], label="MACD")
    plt.plot(tech["date"], tech["Signal"], label="Signal")

    plt.title(f"{stock_name} MACD")
    plt.xlabel("Date")
    plt.ylabel("MACD")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

db.close()
print("The database connection has been closed!")
```
